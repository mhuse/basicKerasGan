```python
paths = list()
for root, dirs, files in os.walk('./dataset'):
     for file in files:
            paths.append(os.path.join(root, file))
            
```

```python
import random
```

```python
def read_batch(paths, batch_size):
    images = list()
    rand_paths = random.choices(paths, k=batch_size)
    for img_path in rand_paths:
        images.append(cv2.imread(img_path))
    return np.array(images)
    
```

```python
batch = read_batch(paths, 12)
```

```python
import matplotlib.pyplot as plt
```

```python
_, axs = plt.subplots(3, 4, figsize=(12, 12))
axs = axs.flatten()
for img, ax in zip(batch, axs):
    ax.imshow(img)
plt.show()
```
